```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_excel("C:\\Users\\USER DK\Downloads\\df.xlsx")
df.head()
```

```python
df.info()
```

```python
df['ship_date']= pd.to_datetime(df['ship_date'])
df['order_date']= pd.to_datetime(df['order_date'])
df.info()
```

```python
df['day']=df['order_date'].dt.day_name()
df['month']=df['order_date'].dt.month_name()
df['month_num']=df['order_date'].dt.month
df['year']=df['order_date'].dt.year
df['ship_delay']=df['ship_date']-df['order_date']
df['ship_delay']=df['ship_delay'].dt.days
df.head()
```

```python
df_grouped = df.groupby(['city'], as_index=False)['profit'].sum().sort_values(by='profit',ascending=[False])
df_grouped = df_grouped.reset_index(drop=True)
df_grouped
```

```python
import matplotlib.pyplot as plt
import io
```
